# Función para carga de datos

```python
def process_data(df):
    df = df.drop_duplicates()

    df['Fecha'] = df['Fecha'].apply(lambda x: pd.Timestamp(x))
    df['datetime'] = pd.to_datetime(df['datetime'])

    df.rename(
        columns={
            'Código universal de punto de suministro': 'CUPS',
            'Fecha': 'fecha',
            'Hora': 'hora',
            'Consumo': 'consumo',
            'Método de obtención': 'met_obtencion'
        }, inplace=True)

    return df
```

```python
df = pd.read_csv("data/electrodatos.csv")
df = process_data(df)
df.to_csv("data/electrodatos_processed.csv", index=False)
df
```
